# 기말 프로젝트 안내

- 앞서 제출한 나만의 웹앱 기획서에 따라, 데이터를 수집·정제·분석·시각화하여 최종적으로 Streamlit 웹 앱 형태로 배포하시오.
- 12~13주차 학습 내용 중 하나인 BeautifulSoup 또는 Selenium을 활용한 데이터 수집 과정이 반드시 포함되어야 한다. (단, 웹 앱 내에서 실시간으로 데이터를 수집할 필요는 없음.)
- 웹 앱에서는 데이터에 대한 설명, 분석 방법과 분석 결과에 대한 설명이 반드시 포함되어야 한다.

      - 중간 프로젝트에서 이미 제출한 경우라도 업데이트 하여 다시 제출하셔도 됩니다.
